```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('cars.csv')
```

```python
df['owner'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,0:4],df.iloc[:,4],test_size=0.2,random_state=12)
```

```python
x_train,y_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
transformer=ColumnTransformer(transformers=[
    ('tf1',OrdinalEncoder(categories=[['First Owner','Second Owner','Third Owner','Fourth & Above Owner','Test Drive Car']]),['owner']),
    ('tf2',OneHotEncoder(sparse=False,drop='first',dtype=int),['fuel'])
],remainder='passthrough')
```

```python
data=transformer.fit_transform(x_train)
```
